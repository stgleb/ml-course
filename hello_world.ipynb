Now, lets draw some plots

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate normally distributed data
mean = 0      # Center of the distribution
std_dev = 1   # Standard deviation (spread)
num_samples = 1000

data = np.random.normal(loc=mean, scale=std_dev, size=num_samples)

# Step 2: Plot the distribution
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, edgecolor='black', density=True)
plt.title('Normal Distribution (mean=0, std=1)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

Now lets generae uniformal distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate uniformly distributed data
num_samples = 100
x = np.linspace(0, 10, num_samples)
y = np.random.uniform(low=0, high=1, size=num_samples)

# Step 2: Plot the line and fill under it
plt.figure(figsize=(10, 6))
plt.plot(x, y, label="Uniform Distribution", color='black')  # Line
plt.fill_between(x, y, color='blue', alpha=0.3)               # Fill under line

plt.title('Uniform Distribution with Fill')
plt.xlabel('X-axis')
plt.ylabel('Y-axis (Uniform Samples)')
plt.legend()
plt.grid(True)
plt.show()
```
